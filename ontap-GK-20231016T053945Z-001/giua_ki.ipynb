```python
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
# Tham só phân phối chuẩn

mean = 172;
std_dev = 5;

# a) Xác suất chọn ngẫu nhiên một bạn gọi là lùn (nhỏ hơn 162 cm)
cau_a = st.norm.cdf(162, loc=mean, scale=std_dev).round(2)
print("Câu a: ", cau_a)

# b) Xác suất chọn ngẫu nhiên 1 bạn gọi là cao (lớn hơn 182cm)
cau_b = (1-st.norm.cdf(182, loc=mean, scale=std_dev)).round(2)
print("Câu b: ", cau_b)

# c) Xác suất chọn người tầm thước (trong khoảng 165-182(cm))
cau_c = (st.norm.cdf(182, loc=mean, scale=std_dev) - st.norm.cdf(165, loc=mean, scale=std_dev)).round(2);
print(cau_c) 

# d) Chuẩn thấp bé của thanh niên ở Hàn quốc
cau_d = st.norm.ppf(0.06, loc=mean, scale=std_dev).round(2)
print(cau_d)

# e) Chuẩn cao lớn của thanh niên ở Hàn quốc
cau_e = st.norm.ppf(0.9, loc=mean, scale=std_dev).round(2)
print(cau_e)

x = range(150, 195);
pdf = st.norm.pdf(x, loc=mean, scale=std_dev)

plt.plot(x, pdf)
```
