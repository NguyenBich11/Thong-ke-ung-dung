### Câu 1:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.title('First Graph!')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

x = np.linspace(0, 2.6, 50)
plt.plot(x, x*x,'red', label="Y=X**2")
plt.legend(loc= "upper left")
x = np.linspace(2.6, 4, 100)
plt.plot(x, x*x,'r--')
plt.legend(loc= "upper left")
x = np.linspace(0, 4, 5)
plt.plot(x, 2*x,'b--', marker='^', label="Y=2X")
plt.legend(loc="upper left")
plt.show()
```

### Câu 2:

```python
plt.subplot(1, 2, 1)
x = np.linspace(0, 10, 10)
plt.plot(x, 2*x,'b--',marker='.', label="y = 2*x")
plt.legend(loc="upper left")

plt.subplot(1,2,2)
x = np.linspace(0, 10, 10)
plt.plot(x, x*x,'green',marker='*',mfc='green', label="y = x*x")
plt.legend(loc="upper left")
plt.show()
```

### Câu 3:

```python
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
x = np.linspace(0, 10, 10)
plt.plot(x, x**2,'blue',marker='o',mfc='red', label="y = x**2")
plt.legend(loc="upper left")

plt.plot(x, x**3,'orange',marker='*',mfc ='black', mec='black', label="y = 2*x")
plt.legend(loc="upper left")
plt.show()
```

### Câu 4:

```python
plt.title('Coffee Consumption')
plt.ylabel('Numbers of cups')
last_week_cups = (20, 35, 30, 35, 27)
this_week_cups = (25, 32, 34, 20, 25)
names = ['Mary', 'Paul', 'Billy', 'Franka', 'Stephan']
barWidth = 0.35
br1 = np.arange(len(last_week_cups))
br2 = [x + barWidth for x in br1]
plt.bar(br1,last_week_cups,color='blue',width = barWidth, label ='Last week')
plt.bar( br2,this_week_cups,color='orange',width = barWidth, label ='This week')
plt.legend(loc="upper right")
plt.xlabel('Branch', fontweight = 'bold', fontsize=10)
plt.xticks([r + barWidth/2 for r in range(len(last_week_cups))],names)
plt.show()
```

### Câu 5:

Vẽ đồ thị tán xạ (scatter plot) với dữ liệu được phát sinh nhẫu nhiên như dưới đây. Bạn hãy đưa ra nhận xét của mình.

```python
x = np.linspace(0, 100, 100)
y = x + np.random.randn(len(x))*5
z = x + np.random.randn(len(x))*20
```

```python

# Vẽ đồ thị
plt.scatter(x, y, label='random error in [0, 5]', color = 'purple' )
plt.scatter(x, z, label='random error in [0, 20]',color = 'orange')
plt.plot([0, 100], [0, 100], color = 'red')

# Đặt tên cho trục x, y và tiêu đề đồ thị
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.title('Mô phỏng đồ thị y = x với các sai số khác nhau')
# Hiển thị chú thích
plt.legend()

# Hiển thị đồ thị
plt.show()
```
