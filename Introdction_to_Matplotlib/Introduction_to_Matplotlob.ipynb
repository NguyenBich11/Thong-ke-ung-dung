##Pyploy API

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)")
plt.plot(x, np.cos(x), color="red", label="cos(x)")
plt.xlabel("Bien X")
plt.ylabel("Sin(X)")
##plt.xlim([0, 4])
##plt.axis([0, 4, 0, 6])
plt.axis("tight")
plt.axis("equal")

plt.legend();
```

## Object-Oriented API

```python
#1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

#3. Plot the data
ax.plot(x, y);
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```
